```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index('day', inplace=True) #inplace Ttue modyfies the original df
df
```

```python
df.shape
```

```python
new_df = df.fillna(0)
new_df
```

```python
"""
define how you want to fill the NaN
"""

new_df = df.fillna({
    'temperature' :0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
"""
ffill = forward fill
method = "ffill" carries over whatever the previous value to NaN
"""
new_df = df.fillna(method="ffill") 
new_df
```

```python
new_df = df.fillna(method="ffill", limit = 1)# limits how many NaN can be filled
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method = "time") #choose the nearest time of 2
new_df
```

```python
new_df = df.dropna(how = "all") #drop the row that has all NaN
new_df
```

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(idx)
new_df
```
